# Fully Connected Model

The fully connected model is composed of dense layers + convolutional layers.

Given the similarities of this dataset with the MNIST dataset, I choose to build a similar network architecture used to generate digits from the MNIST dataset as shown in this article: [How to Develop a GAN for Generating MNIST Handwritten Digits]("https://machinelearningmastery.com/how-to-develop-a-generative-adversarial-network-for-an-mnist-handwritten-digits-from-scratch-in-keras/")

## Defining The Discriminator

```python
discriminator = keras.Sequential(
    [
        keras.Input(shape=(32, 32, 1)),
        layers.Conv2D(64, kernel_size=3, strides=2, padding="same"),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(64, kernel_size=3, strides=2, padding="same"),
        layers.LeakyReLU(alpha=0.2),
        layers.Flatten(),
        layers.Dropout(0.4),
        layers.Dense(1, activation="sigmoid"),
    ],
    name="discriminator",
)
discriminator.summary()
```

## Defining the generator

```python
latent_dim = 128

generator = keras.Sequential(
    [
        keras.Input(shape=(latent_dim,)),
        layers.Dense(8 * 8 * 128),
        layers.Reshape((8, 8, 128)),
        layers.Conv2DTranspose(128, kernel_size=4, strides=2, padding="same"),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2DTranspose(128, kernel_size=4, strides=2, padding="same"),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(1, kernel_size=5, padding="same", activation="sigmoid"),
    ],
    name="generator",
)
generator.summary()
```

## Defining The Model's Class

```python
class GAN(keras.Model):
    def __init__(self, discriminator, generator, latent_dim):
        super().__init__()
        self.discriminator = discriminator
        self.generator = generator
        self.latent_dim = latent_dim

    def compile(self, d_optimizer, g_optimizer, loss_fn):
        super().compile()
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer
        self.loss_fn = loss_fn
        self.d_loss_metric = keras.metrics.Mean(name="d_loss")
        self.g_loss_metric = keras.metrics.Mean(name="g_loss")

    @property
    def metrics(self):
        return [self.d_loss_metric, self.g_loss_metric]

    def train_step(self, real_images):
        # Sample random points in the latent space
        batch_size = tf.shape(real_images)[0]
        random_latent_vectors = tf.random.normal(shape=(batch_size, self.latent_dim))

        # Decode them to fake images
        generated_images = self.generator(random_latent_vectors)

        # Combine them with real images
        combined_images = tf.concat([generated_images, real_images], axis=0)

        # Assemble labels discriminating real from fake images
        labels = tf.concat(
            [tf.ones((batch_size, 1)), tf.zeros((batch_size, 1))], axis=0
        )
        # Add random noise to the labels - important trick!
        # adding this noise helps with the stability by making
        # the job of the discriminator harder, this reduces chances
        # of overfitting.
        labels += 0.05 * tf.random.uniform(tf.shape(labels))

        # Train the discriminator
        with tf.GradientTape() as tape:
            predictions = self.discriminator(combined_images)
            d_loss = self.loss_fn(labels, predictions)
        grads = tape.gradient(d_loss, self.discriminator.trainable_weights)
        self.d_optimizer.apply_gradients(
            zip(grads, self.discriminator.trainable_weights)
        )

        # Sample random points in the latent space
        random_latent_vectors = tf.random.normal(shape=(batch_size, self.latent_dim))

        # Assemble labels that say "all real images"
        misleading_labels = tf.zeros((batch_size, 1))

        # Train the generator (note that we should *not* update the weights
        # of the discriminator)!
        with tf.GradientTape() as tape:
            predictions = self.discriminator(self.generator(random_latent_vectors))
            g_loss = self.loss_fn(misleading_labels, predictions)
        grads = tape.gradient(g_loss, self.generator.trainable_weights)
        self.g_optimizer.apply_gradients(zip(grads, self.generator.trainable_weights))

        # Update metrics
        self.d_loss_metric.update_state(d_loss)
        self.g_loss_metric.update_state(g_loss)
        return {
            "d_loss": self.d_loss_metric.result(),
            "g_loss": self.g_loss_metric.result(),
        }
```

The model is not trained yet but we can generate images
with some random latent vector.

```python
model = GAN(discriminator, generator, latent_dim)
random_latent_vectors = tf.random.normal(shape=(10, latent_dim))
generated_images = model.generator(random_latent_vectors)
plot_images(generated_images)
```

```python
# load a trained model if it exists

latest_weights = tf.train.latest_checkpoint('trained_models/gan_fully_connected_no_regularization')

if latest_weights is not None:
    print('Loading model weights')
    model = GAN(discriminator, generator, latent_dim)
    model.load_weights(latest_weights)
    random_latent_vectors = tf.random.normal(shape=(10, latent_dim))
    generated_images = model.generator(random_latent_vectors)
else:
    print('Starting training')
    epochs = 100  
    
    gan = GAN(discriminator=discriminator, generator=generator, latent_dim=latent_dim)
    gan.compile(
        d_optimizer=keras.optimizers.Adam(learning_rate=0.0001),
        g_optimizer=keras.optimizers.Adam(learning_rate=0.0001),
        loss_fn=keras.losses.BinaryCrossentropy(),
    )
    
    gan.fit(resized_train_data.astype('float32'), epochs=epochs)
    model = GAN(discriminator, generator, latent_dim)

    # Saving the trained model weights
    # import os.path
    # # os.makedirs('gan_fully_connected_no_regularization', exist_ok=True)
    # if os.path.isfile('trained_models/gan_fully_connected_no_regularization/gan_fully_connected_no_regularization_weights') is False:
    #     model.save_weights('trained_models/gan_fully_connected_no_regularization/gan_fully_connected_no_regularization_weights', save_format="tf")
```

## Latent Space

When using GAN's to generate images, we want to be able to have control over the resulting images given an input latent vector,  to achieve this we must assert that the latent space is regular. A regularised latent space means that if we take a small area in the latent space and use these points as inputs to generate images, the expected result should be that all the generated images are similar and meaningful.

To visually check if our latent space is regularised we can generate images using as input a random point from the latent space and its neighbors.

```python
## Extract the a point from the latent space
base_random_input = tf.random.normal(shape=(1, latent_dim))
```

Aproach 1: Generate new latent vectors by adding small amount of rando noises to the base_random_input.

```python
## Add a small noise to the initial point, generating 10 new points
random_latent_vectors =  base_random_input + 0.2*tf.random.normal(shape=(10, latent_dim))
generated_images = model.generator(random_latent_vectors)
plot_images(generated_images)
```

Aproach 2: Generate new latent vectors by continuously adding a small value to the initial point.

```python
## Continuously add a small value to the initial point to 
## Continuously move through the latent space
random_latent_vectors = tf.stack([(np.mean(base_random_input)*3 + base_random_input[0])*i for i in range(1,11) ])
generated_images = model.generator(random_latent_vectors)
plot_images(generated_images)
```

We can see by these samples that a small perturbation in the `base_random_input` lead to small changes in the resulting images, suggesting that the latent space is regular.

# Fully Convolutional Model

In the fully convolutional model we will replace the dense layer in the generator input, so the latent space input
will be a NxN vector

## Defining The Discriminator

```python
discriminator = keras.Sequential(
    [
        keras.Input(shape=(32, 32, 1)),
        layers.Conv2D(128, kernel_size=3, strides=2, padding="same"),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(64, kernel_size=3, strides=2, padding="same"),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(1, kernel_size=3, strides=8, activation='sigmoid', padding="same"),
        layers.Dropout(0.2),
        layers.Reshape([1])
    ],
    name="discriminator",
)
discriminator.summary()
```

## Defining The Generator

```python
latent_shape = (8,8)

generator = keras.Sequential(
    [
        keras.Input(shape=(*latent_shape, 1)),
        layers.Conv2DTranspose(128*4, kernel_size=4, strides=1, padding="same"),
        layers.LeakyReLU(alpha=0.2),     
        layers.Conv2DTranspose(128*3, kernel_size=4, strides=2, padding="same"),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2DTranspose(128*2, kernel_size=4, strides=2, padding="same"),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(1, kernel_size=5, padding="same", activation="sigmoid"),
    ],
    name="generator",
)
generator.summary()
```

## Defining The Model's class

```python
class FCVGAN(keras.Model):
    def __init__(self, discriminator, generator, latent_shape):
        super().__init__()
        self.discriminator = discriminator
        self.generator = generator
        self.latent_shape = latent_shape

    def compile(self, d_optimizer, g_optimizer, loss_fn):
        super().compile()
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer
        self.loss_fn = loss_fn
        self.d_loss_metric = keras.metrics.Mean(name="d_loss")
        self.g_loss_metric = keras.metrics.Mean(name="g_loss")

    @property
    def metrics(self):
        return [self.d_loss_metric, self.g_loss_metric]

    def train_step(self, real_images):
        # Sample random points in the latent space
        batch_size = tf.shape(real_images)[0]
        random_latent_vectors = tf.random.normal(shape=(batch_size, *self.latent_shape))

        # Decode them to fake images
        generated_images = self.generator(random_latent_vectors)

        # Combine them with real images
        combined_images = tf.concat([generated_images, real_images], axis=0)

        # Assemble labels discriminating real from fake images
        labels = tf.concat(
            [tf.ones((batch_size, 1)), tf.zeros((batch_size, 1))], axis=0
        )
        # Add random noise to the labels - important trick!
        # adding this noise helps with the stability by making
        # the job of the discriminator harder, this reduces chances
        # of overfitting.
        labels += 0.05 * tf.random.uniform(tf.shape(labels))

        # Train the discriminator
        with tf.GradientTape() as tape:
            predictions = self.discriminator(combined_images)
            d_loss = self.loss_fn(labels, predictions)
        grads = tape.gradient(d_loss, self.discriminator.trainable_weights)
        self.d_optimizer.apply_gradients(
            zip(grads, self.discriminator.trainable_weights)
        )

        # Sample random points in the latent space
        random_latent_vectors = tf.random.normal(shape=(batch_size, *self.latent_shape))

        # Assemble labels that say "all real images"
        misleading_labels = tf.zeros((batch_size, 1))

        # Train the generator (note that we should *not* update the weights
        # of the discriminator)!
        with tf.GradientTape() as tape:
            predictions = self.discriminator(self.generator(random_latent_vectors))
            g_loss = self.loss_fn(misleading_labels, predictions)
        grads = tape.gradient(g_loss, self.generator.trainable_weights)
        self.g_optimizer.apply_gradients(zip(grads, self.generator.trainable_weights))

        # Update metrics
        self.d_loss_metric.update_state(d_loss)
        self.g_loss_metric.update_state(g_loss)
        return {
            "d_loss": self.d_loss_metric.result(),
            "g_loss": self.g_loss_metric.result(),
        }
```

# Discussion

This test aimed to use GANs (Generative Adversarial Neural Networks) to generate meander channel images. The dataset consisted of 50,000 images, in the preporcess step the images were resized from 128x128 pixels to 32x32 pixels. Two types of GANs were implemented: a fully connected model and a fully convolutional model.

The fully connected model had a latent space of size 128 dimensions and underwent training on the 50,000 images for 100 epochs. Due to the similarity between the meander dataset and the well-known MNIST dataset, a proven architecture for the MNIST was chosen. The model successfully generated good-quality meander images with minimal artifacts, resulting in clear contrast between the channel and non-channel regions. Analysis of the latent space suggested regularization occurred during the training process.

Transforming the fully connected model into a fully convolutional model proved more challenging as there was no initial starting point for a suitable architecture. In this transformation, the generator's input was changed to a (16,16,1) tensor, and the discriminator's last layer was modified to a Conv2D with only one filter, adopting a sigmoid activation. The number of strides was adjusted to yield a (1,1,1) tensor, and a reshape layer was added to output a scalar value. Multiple configurations were tested to maintain a similar number of trainable parameters as the fully connected model. However, satisfactory results could not be achieved through training.

While the fully connected model achieved the desired results, further computational power would likely improve its performance. On the other hand, the fully convolutional model failed to reach the same level as the fully connected model. Additional experimentation is necessary to fine-tune this architecture.
