# Variational AutoEnconder - Fully Connected

## Defining Sampling Class

The sampling class introduces regularization in the latent space of the autoencoder by sampling a latent vector from the mean (z_mean) and variance (z_log_var). It is incorporated as the last layer in the encoder, which returns the sampled vector along with the mean and variance.
These statistical properties are then utilized to calculate the kl_loss (Kullback–Leibler loss), which measures the statistical distance between the latent vector distribution and a normal distribution. The kl_loss is combined with the reconstruction loss, thereby compelling the encoder to generate a latent space where input images are distributed according to Gaussian distributions. Is important to notice that the samplig class do not calculate the mean and the variance explicitly, insteat it guides the model to return this values during the training.

```python
class Sampling(layers.Layer):
    """Uses (z_mean, z_log_var) to sample z, the vector encoding a meander."""

    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon
```

## Defining The Encoder

```python
# The latent space dimention is the same as in the GAN model.
latent_dim = 128

encoder_inputs = keras.Input(shape=(32, 32, 1))
x = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(encoder_inputs)
x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Flatten()(x)
x = layers.Dense(1000, activation="relu")(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()
```

## Defining The Decoder

```python
latent_inputs = keras.Input(shape=(latent_dim,))
x = layers.Dense(8 * 8 * 64, activation="relu")(latent_inputs)
x = layers.Reshape((8, 8, 64))(x)
x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
decoder_outputs = layers.Conv2DTranspose(1, 3, activation="sigmoid", padding="same")(x)
decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()
```

## Defining The Mode'l Class

```python
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(
            name="reconstruction_loss"
        )
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    keras.losses.binary_crossentropy(data, reconstruction), axis=(1, 2)
                )
            )
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))
            total_loss = reconstruction_loss + kl_loss
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }
```

## Training The Model

```python
latest_weights = tf.train.latest_checkpoint('trained_models/vae_fully_connected')
if latest_weights is not None:
    print('Loading model weights')
    model = VAE(encoder, decoder)
    model.load_weights(latest_weights)
else:
    print('Starting training')
    model = VAE(encoder, decoder)
    model.compile(optimizer=keras.optimizers.Adam())
    model.fit(resized_train_data[:20000].astype('float32'), epochs=30, batch_size=128)

    # Saving the trained model weights
    import os.path
    os.makedirs('vae_fully_connected', exist_ok=True)
    if os.path.isfile('trained_models/vae_fully_connected') is False:
        model.save_weights('trained_models/vae_fully_connected/vae_fully_connected_weights', save_format="tf")
```

```python
random_latent_vectors = tf.random.normal(shape=(10, latent_dim))
generated_images = model.decoder.predict(random_latent_vectors)
generated_images = tf.squeeze(generated_images)
plot_images(generated_images)
```

## Latent Space

To visually check if our latent space is regularised we can generate images using as input a random point from the latent space and its neighbors.

```python
base_random_input = tf.random.normal(shape=(1, latent_dim))
random_latent_vectors =  base_random_input + 0.1*tf.random.normal(shape=(10, latent_dim))
generated_images = model.decoder.predict(random_latent_vectors)
generated_images = tf.squeeze(generated_images)
plot_images(generated_images)
```

# Variational AutoEnconder - Fully Convolutional

## Defining Sampling Class

```python
class Sampling(layers.Layer):
    """Uses (z_mean, z_log_var) to sample z, the vector encoding a meander."""

    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        epsilon = tf.keras.backend.random_normal(shape=(batch, 8, 8, 1))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon
```

## Defining The Encoder

```python
# The latent space dimantion is the same as in the GAN model.
latent_shape = (8,8)

encoder_inputs = keras.Input(shape=(32, 32, 1))
x = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(encoder_inputs)
x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2D(1000, 3, activation="relu", strides=1, padding="same")(x)
z_mean = layers.Conv2D(1, 3, activation="relu", strides=1, padding="same")(x)
z_log_var = layers.Conv2D(1, 3, activation="relu", strides=1, padding="same")(x)
z = Sampling()([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()
```

## Defining The Decoder

```python
latent_inputs = keras.Input(shape=(*latent_shape,1))
x = layers.Conv2DTranspose(8*8*64, 1, activation="relu", strides=2, padding="same")(latent_inputs)
x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(32, 3, activation="relu", strides=1, padding="same")(x)
decoder_outputs = layers.Conv2DTranspose(1, 1, activation="sigmoid", padding="same")(x)
decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()
```

## Defining The Mode'l Class

```python
class FCVAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(
            name="reconstruction_loss"
        )
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    keras.losses.binary_crossentropy(data, reconstruction), axis=(1, 2)
                )
            )
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))
            total_loss = reconstruction_loss + kl_loss
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }
```

```python
latest_weights = tf.train.latest_checkpoint('trained_models/vae_fully_convolutional')
if latest_weights is not None:
    print('Loading model weights')
    model = FCVAE(encoder, decoder)
    model.load_weights(latest_weights)
else:
    print('Starting training')
    model = FCVAE(encoder, decoder)
    model.compile(optimizer=keras.optimizers.Adam())
    model.fit(resized_train_data[:2000].astype('float32'), epochs=30, batch_size=128)

    # Saving the trained model weights
    import os.path
    os.makedirs('vae_fully_convolutional', exist_ok=True)
    if os.path.isfile('trained_models/vae_fully_convolutional') is False:
        model.save_weights('trained_models/vae_fully_connected/vae_fully_convolutional_weights', save_format="tf")
```

```python
random_latent_vectors = tf.random.normal(shape=(10, *latent_shape))
generated_images = model.decoder.predict(random_latent_vectors)
generated_images = tf.squeeze(generated_images)
plot_images(generated_images)
```

# Discussion

This test aimed to utilize Variational Autoencoders (VAEs) for generating meander channel images. The dataset comprised 50,000 images, which were resized from 128x128 pixels to 32x32 pixels during the preprocessing step. Two types of VAEs were implemented: a fully connected model and a fully convolutional model.

The fully connected model was trained on 20,000 images, and for the purpose of comparison, its latent space dimension was set to be the same as the GAN models (128-dimensional space). During training, the latent space was regularized by incorporating a Kullback–Leibler loss term in the total loss. To inspect the regularization of the latent space, a point in the latent space was perturbed from its initial value. The resulting generated images exhibited good contrast between channel and non-channel regions, but the shapes of the channels remained distorted. Since the size of the decoder was smaller than the encoder, one possible improvement could be to increase the decoder size and train it on the full 50,000 images instead of just 20,000.

The fully convolutional model was trained on 2,000 images due to its larger size. To transform the fully connected model into a fully convolutional one, I modified the output of the encoder to be a (8,8,1) tensor. Additionally, the sampling class was adapted to generate latent vectors of this shape. Finding the appropriate structure for this model proved challenging. Unfortunately, the resulting images were of lower quality compared to the fully connected model.
